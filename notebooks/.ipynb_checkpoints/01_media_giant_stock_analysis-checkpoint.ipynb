# Media Giants Stock Analysis

## Introduction
This project analyzes the stock performance and volatility of four major global entertainment companies: Disney, Comcast (Universal), Warner Bros Discovery, and Sony. These companies span theme parks, film studios, streaming platforms, gaming, and consumer entertainment. The goal is to understand how diversified media giants respond to market shifts, tourism cycles, and evolving audience behavior.

## Setup & Download
Import libraries, define tickers, download adjusted close prices.

### Import Libraries

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Define Tickers

```python
tickers = ["DIS", "CMCSA", "WBD", "SONY"]
sns.set(style='whitegrid')
```

### Download Daily Adjusted Close Prices

```python
data = yf.download(
    tickers,
    start="2019-01-01",
    end=None,
    interval="1d",
    auto_adjust=True,
    threads=False
)

close_prices = data["Close"]
close_prices.head()
```

## Data Cleaning & Preparation
Ensure the dataset is consistent and ready for analysis.
- Handling missing values
- Calculating daily returns
- Calculating cumulative returns

### Handle Missing Values

```python
close_prices = close_prices.dropna()
```

### Create Returns Dataset

```python
returns = close_prices.pct_change().dropna()
returns.head()
```

### Calculate Cumulative Returns

```python
cumulative_returns = (1 + returns).cumprod()
cumulative_returns
```

## Exploratory Data Analysis (EDA)
Visuals:
- Price History
- Cumulative Returns
- Distribution of Daily Returns
- Correlation Heatmap

### Stock Price History Plot

```python
plt.figure(figsize=(12,6))
for t in tickers:
    plt.plot(close_prices.index, close_prices[t], label=t)

plt.title("Stock Price History (2019-Present)")
plt.xlabel("Date")
plt.ylabel("Adjusted Close Price (USD)")
plt.legend()

plt.savefig("../outputs/stock_price_history.png", dpi=300, bbox_inches="tight")
plt.show()
```

This plot shows how each company's stock price has evolved over time. It highlights major market events, including the sharp decline during the COVID-19 shutdowns and the varying recovery trajectories afterward. The differences in slope and voltality reflect how each company's business model responds to industry shifts, tourism cycles, and content performence.

### Cumulative Returns Plot

```python
plt.figure(figsize=(12,6))
for t in tickers:
    plt.plot(cumulative_returns.index, cumulative_returns[t], label=t)

plt.title("Cumulative Returns (2019-Present)")
plt.xlabel("Date")
plt.ylabel("Growth of $1 Investment")
plt.legend()

plt.savefig("../outputs/cumulative_returns.png", dpi=300, bbox_inches="tight")
plt.show()
```

Cumulative returns illustrate how a $1 investment in each company would have grown over the analysis period. This makes longâ€‘term performance differences easy to compare. Companies with more stable or diversified revenue streams show smoother, more consistent growth, while those facing restructuring or industry pressure display more pronounced swings.

### Distribution of Daily Returns

```python
plt.figure(figsize=(12,6))

for t in tickers:
    sns.kdeplot(
        returns[t],
        fill=True,
        alpha=0.3,
        linewidth=1.5,
        label=t
    )

plt.title("Distribution of Daily Returns (2019-Present)", fontsize=16, weight="bold")
plt.xlabel("Daily Return", fontsize=13)
plt.ylabel("Density", fontsize=13)
plt.xlim(-0.1, 0.1)

plt.legend()
plt.tight_layout()
plt.savefig("../outputs/distribution_of_daily_returns.png", dpi=300, bbox_inches="tight")
plt.show()
```

The distribution of daily returns reveals how frequently each company experiences small versus large price movements. Wider or heavierâ€‘tailed distributions indicate higher volatility, while tighter curves suggest more stable dayâ€‘toâ€‘day behavior. This helps identify which companies carry more shortâ€‘term risk and which tend to move more predictably.

### Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.set_style("white")

corr=returns.corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    linecolor="white",
    square=True,
    annot_kws={"size":12}
)

plt.title("Correlation of Daily Returns", fontsize=16, weight="bold")
plt.tight_layout()
plt.savefig("../outputs/correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

The correlation heatmap shows how similarly the companiesâ€™ daily returns move relative to one another. Higher correlations suggest shared industry drivers or similar exposure to market conditions. Lower correlations indicate more distinct business models or revenue sources. This visualization helps explain why some companies behave alike while others diverge.

## Risk & Return Analysis
Metrics:
- Annualized Returns
- Volatiltiy
- Risk vs Return Scatterplot

### Annualized Returns
This is the mean daily return multiplied by 252 trading days.

```python
annualized_returns = returns.mean() * 252
annualized_returns.to_frame("Annualized Return")
```

### Volatility
This is the standard deviation of daily returns, scaled to a year.

```python
volatility = returns.std() * np.sqrt(252)
volatility.to_frame("Volatility")
```

## Company-Level Insights

### COVID Recovery
    Disney and Comcast show the clearest COVIDâ€‘related patterns because of their theme park exposure. Both experienced sharp declines during shutdowns, followed by steady recovery as parks reopened and tourism rebounded. Sony recovered more quickly due to strong gaming demand during lockdowns, which insulated it from the worst of the downturn. Warner Bros Discovery had the rockiest path, with recovery slowed by the Warnerâ€“Discovery merger, restructuring costs, and challenges in the linear TV market.

### Streaming vs. Park-Heavy Companies
    Disney and Comcast balance parks with streaming, creating mixed performance patterns. Parks drive strong rebounds after COVID, while streaming introduces volatility tied to subscriber growth and profitability concerns. Sony stands apart: its gaming division provides recurring revenue and stability, making it less sensitive to streaming trends. Warner Bros Discovery is the most streamingâ€‘dependent, and its stock reflects the uncertainty of competing in a crowded streaming market.

### Diversified Revenue Streams Affect on Volatility
    Sony and Comcast benefit the most from diversification. Sonyâ€™s combination of gaming, music, film, and electronics smooths out volatility across market cycles. Comcastâ€™s broadband and cable segments act as stabilizers when media segments fluctuate. Disney is moderately diversified but still sensitive to tourism and content cycles. Warner Bros Discovery has the least stabilizing diversification, with heavy exposure to streaming and linear TV, resulting in the highest volatility of the group.

### Strongest Long-Term Performance
    Sony demonstrates the strongest longâ€‘term performance, supported by consistent PlayStation revenue, strong IP, and global entertainment assets. Comcast follows with steady, defensible growth driven by broadband and Universalâ€™s strong park performance. Disney shows solid longâ€‘term potential but with more pronounced swings tied to parks and streaming strategy shifts. Warner Bros Discovery trails the group due to mergerâ€‘related uncertainty and the challenges of restructuring in a rapidly changing media landscape.
